```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
## pie chart - Top 3 Countrys That uses Zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct='%3.7f%%')
```

Observation:Zomato maximum records or transections are from India
After that USA and then United Kingdoms

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

```python
final_df.columns
```

## Observation
1. When the rating is between 4.5 to 4.9 that means---->Excellent
2. When the rating is between 4.0 to 3.4 that means----> very good
3. When the rating is between 3.5 to 3.9 that means----> good
4. When the rating is between 3.0 to 3.4 that means----> Average 
5. when the rating is between 2.5 to 2.9 that means----> Average
6. when the rating is between 2.0 to 2.4 that means----> poor
